### MONTHLY PREDICITONS"

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Train 12 separate Random Forest regressors, one for each month
month_regressors = {}
for month_abbr in ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']:
    # Extract the target variable for this month
    y_month = data[month_abbr]
    X_train, X_test, y_train, y_test = train_test_split(X, y_month, test_size=0.2, random_state=42)
    
    # Train the Random Forest regressor for this month
    regressor = RandomForestRegressor(n_estimators=10, random_state=42)
    regressor.fit(X_train, y_train)
    
    # Store the regressor for this month
    month_regressors[month_abbr] = regressor

# User input process
input_year = int(input("Enter the year: "))
input_subdivision_encoded = int(input("Enter the encoded subdivision: "))
input_month_abbr = input("Enter the month abbreviation (e.g., JAN, FEB, ...): ").upper()

# Create user input DataFrame
user_input = pd.DataFrame({
    'YEAR': [input_year],
    'JAN': [mean_values['JAN']],  # You can use mean values or any other method to fill missing values for other months
    'FEB': [mean_values['FEB']],
    'MAR': [mean_values['MAR']],
    'APR': [mean_values['APR']],
    'MAY': [mean_values['MAY']],
    'JUN': [mean_values['JUN']],
    'JUL': [mean_values['JUL']],
    'AUG': [mean_values['AUG']],
    'SEP': [mean_values['SEP']],
    'OCT': [mean_values['OCT']],
    'NOV': [mean_values['NOV']],
    'DEC': [mean_values['DEC']],
    'SUBDIVISION_encoded': [input_subdivision_encoded]
})

# Predict the rainfall for the specified month
predicted_rainfall = month_regressors[input_month_abbr].predict(user_input)

# Calculate metrics
y_pred = month_regressors[input_month_abbr].predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

# Print prediction and metrics
print(f"Predicted rainfall for month {input_month_abbr}: {predicted_rainfall[0]}")
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
```
